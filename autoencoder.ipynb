```python
import numpy as np
from matplotlib import pyplot
from scipy import ndimage


num_training = 1000
picture_list = []
for picture in range(num_training):
    picture_list.append(pyplot.imread("pics/pic0.gif"))

X_train = np.array(picture_list)
X_val = None
    
```

```python
import tensorflow as tf


class AutoEncoder(object):
    def __init__(self, input_dim, hidden_dim, num_epoch, batch_size):
        self.input_dim = input_dim
        self.X = tf.placeholder(tf.float32, [None, 18, 37, 4])
        self.Y = tf.placeholder(tf.float32, [None, 18, 37, 4])
        self.num_epoch = num_epoch
        self.batch_size = batch_size 
        self.hidden_dim = hidden_dim
        self._model()
       
    

    def _model(self):
        

        with tf.variable_scope("encoder"):
            self.hidden = tf.layers.dense(tf.layers.flatten(self.X), self.hidden_dim, activation= tf.nn.relu, kernel_initializer=tf.glorot_uniform_initializer)
        with tf.variable_scope("decoder"):
            self.output = tf.layers.dense(self.hidden, self.input_dim, activation=tf.nn.relu, kernel_initializer=tf.glorot_normal_initializer)

        self.loss_op = tf.nn.l2_loss(tf.math.subtract(tf.layers.flatten(self.Y), self.output))
        self.train_op = tf.train.AdamOptimizer().minimize(self.loss_op)

    def train(self, sess, X_train, X_val):
        sess.run(tf.global_variables_initializer())

        loss_history = []
        step = 0
        
        print("-" * 5 + " Start training " + "-" * 5)
        for epoch in range(self.num_epoch):
            print("%d" % epoch)
            for i in range(num_training // self.batch_size):
                X_ = X_train[i * self.batch_size: (i + 1) * self.batch_size][:]
                Y_ = X_

                fetches = [self.train_op, self.loss_op]

                feed_dict = {self.X:X_, self.Y:Y_}
                _, loss = sess.run(fetches, feed_dict)
                
                loss_history.append(loss)
                if i % 1000 == 0:
                    print("Loss for epoch %d: %f" % (i, loss))
            #print("Loss for epoch %d: %f" % (i, loss_his[-1]))



                step += 1



        


```

```python
tf.reset_default_graph()
with tf.Session() as sess:
    with tf.device("/cpu:0"):
        model = AutoEncoder(18 * 37 * 4, 200, 20, 1)
        model.train(sess, X_train, X_val)
```
